Lets investigate the data composition,some columns might not be necessary and if there are too many NaN values or misplaced entries we might have to clean the data.

```python
df.isna().sum()
```

The CRASH_RECORD_ID column will be dropped since its string values do not provide any useful information for our analysis. Additionally, the CRASH_DATE_EST_I column contains only NaN values, so this column will be completely removed.

We also need to drop the following columns as they each contain over 800,000 missing entries:

INTERSECTION_RELATED_I
NOT_RIGHT_OF_WAY_I
HIT_AND_RUN_I
PHOTOS_TAKEN_I
STATEMENTS_TAKEN_I
DOORING_I
WORK_ZONE_I
WORK_ZONE_TYPE
WORKERS_PRESENT_I
The following columns share the same 1,953 rows with NaN values, and these rows will also be removed:

MOST_SEVERE_INJURY
INJURIES_TOTAL
INJURIES_FATAL
INJURIES_INCAPACITATING
INJURIES_NON_INCAPACITATING
INJURIES_REPORTED_NOT_EVIDENT
INJURIES_NO_INDICATION
INJURIES_UNKNOWN



```python
#Drop Crash_RECORD_ID and CRASH_DATE_EST_I entirly
df=df.drop(columns=['CRASH_RECORD_ID','CRASH_DATE_EST_I','INTERSECTION_RELATED_I','NOT_RIGHT_OF_WAY_I','HIT_AND_RUN_I','PHOTOS_TAKEN_I','STATEMENTS_TAKEN_I','DOORING_I','WORK_ZONE_I','WORK_ZONE_TYPE','WORKERS_PRESENT_I','LATITUDE','LONGITUDE','LOCATION','LANE_CNT'])

#Drop the rows with NaN values in the second column of columns discuessed above:
df=df.dropna(subset=['MOST_SEVERE_INJURY','INJURIES_TOTAL','INJURIES_FATAL','INJURIES_INCAPACITATING','INJURIES_NON_INCAPACITATING','INJURIES_REPORTED_NOT_EVIDENT','INJURIES_NO_INDICATION','INJURIES_UNKNOWN','REPORT_TYPE','STREET_DIRECTION','STREET_NAME','BEAT_OF_OCCURRENCE'])
df.isna().sum()
```

```python
df.loc[:,'INJURIES_UNKNOWN'].unique()
```

We will remove the CRASH_DATE column, as it includes the month, day, hour, and minute information, which are already captured in the CRASH_MONTH, CRASH_DAY_OF_WEEK, and CRASH_HOUR columns as float values. We also see that the column INJURIES_UNKNOWN contains all 0, hence we will drop this one too since it wont serve our analysis.

```python
df.drop(columns=['CRASH_DATE','INJURIES_UNKNOWN'],inplace=True)
df.head()
```
